## Task 1 - Análisis

### 1. Dimensión espacial de salida

Se considera una imagen de entrada con:

- $W = 256$
- $H = 256$
- $C = 4$

y se utiliza la fórmula espacial:

$$
O = \left\lfloor \frac{W - K + 2P}{S} \right\rfloor + 1
$$

---

#### Opción A: una convolución $7 \times 7$, stride $2$, padding $0$

Para el ancho, se calcula:

$$
W_{out} = \left\lfloor \frac{256 - 7 + 2(0)}{2} \right\rfloor + 1
$$

$$
W_{out} = \left\lfloor \frac{249}{2} \right\rfloor + 1
$$

$$
W_{out} = 124 + 1 = 125
$$

Como la entrada es cuadrada, para la altura se obtiene el mismo resultado:

$$
H_{out} = 125
$$

Por lo tanto, la salida de la **Opción A** es:

$$
\boxed{125 \times 125 \times 64}
$$

---

#### Opción B: tres convoluciones $3 \times 3$, stride $1$, padding $1$, seguidas de Max-Pooling $2 \times 2$, stride $2$

##### Primera convolución $3 \times 3$

Se calcula:

$$
W_{out} = \left\lfloor \frac{256 - 3 + 2(1)}{1} \right\rfloor + 1
$$

$$
W_{out} = \left\lfloor \frac{255}{1} \right\rfloor + 1
$$

$$
W_{out} = 255 + 1 = 256
$$

Entonces, después de la primera convolución, la salida es:

$$
256 \times 256 \times 64
$$

##### Segunda convolución $3 \times 3$

Como se mantiene stride $1$ y padding $1$, la dimensión espacial no cambia:

$$
256 \times 256 \times 64
$$

##### Tercera convolución $3 \times 3$

De nuevo, la dimensión espacial se conserva:

$$
256 \times 256 \times 64
$$

##### Max-Pooling $2 \times 2$, stride $2$

Se aplica la misma fórmula espacial:

$$
W_{out} = \left\lfloor \frac{256 - 2 + 2(0)}{2} \right\rfloor + 1
$$

$$
W_{out} = \left\lfloor \frac{254}{2} \right\rfloor + 1
$$

$$
W_{out} = 127 + 1 = 128
$$

De igual forma:

$$
H_{out} = 128
$$

Por lo tanto, la salida final de la **Opción B** es:

$$
\boxed{128 \times 128 \times 64}
$$

---

### 2. Parámetros entrenables

Para una capa convolucional, la cantidad de pesos se calcula como:

$$
\text{Pesos} = K_h \times K_w \times C_{in} \times C_{out}
$$

> En este cálculo se consideran solo los **pesos**, no los sesgos.  
> Si se quisieran incluir sesgos, se sumaría un sesgo por cada canal de salida.

---

#### Opción A

Se tiene una sola convolución con:

- Kernel: $7 \times 7$
- $C_{in} = 4$
- $C_{out} = 64$

Entonces:

$$
\text{Pesos}_A = 7 \times 7 \times 4 \times 64
$$

$$
\text{Pesos}_A = 49 \times 256 = 12544
$$

Por lo tanto:

$$
\boxed{\text{Opción A} = 12544 \text{ pesos}}
$$

---

#### Opción B

##### Primera convolución

- Kernel: $3 \times 3$
- $C_{in} = 4$
- $C_{out} = 64$

$$
\text{Pesos}_{B1} = 3 \times 3 \times 4 \times 64
$$

$$
\text{Pesos}_{B1} = 9 \times 256 = 2304
$$

##### Segunda convolución

- Kernel: $3 \times 3$
- $C_{in} = 64$
- $C_{out} = 64$

$$
\text{Pesos}_{B2} = 3 \times 3 \times 64 \times 64
$$

$$
\text{Pesos}_{B2} = 9 \times 4096 = 36864
$$

##### Tercera convolución

Tiene exactamente la misma cantidad de pesos:

$$
\text{Pesos}_{B3} = 3 \times 3 \times 64 \times 64
$$

$$
\text{Pesos}_{B3} = 36864
$$

##### Total de la Opción B

$$
\text{Pesos}_B = 2304 + 36864 + 36864
$$

$$
\text{Pesos}_B = 76032
$$

Por lo tanto:

$$
\boxed{\text{Opción B} = 76032 \text{ pesos}}
$$

---

## Comparación directa

- **Opción A:** $12544$ pesos
- **Opción B:** $76032$ pesos

En este caso específico, la **Opción B utiliza más parámetros** que la Opción A.

---

## Justificación con campo receptivo

Desde la teoría del **campo receptivo**, se sabe que apilar convoluciones pequeñas permite que la red observe una región equivalente a un filtro más grande.

### Campo receptivo de la Opción A

Una sola convolución $7 \times 7$ observa directamente una región de:

$$
\boxed{7 \times 7}
$$

---

### Campo receptivo de la Opción B

Cada convolución $3 \times 3$ aumenta gradualmente el campo receptivo efectivo:

- Después de la primera convolución: $3 \times 3$
- Después de la segunda convolución: $5 \times 5$
- Después de la tercera convolución: $7 \times 7$

Es decir, **tres convoluciones $3 \times 3$** logran un campo receptivo equivalente a:

$$
\boxed{7 \times 7}
$$

Además, el max-pooling posterior amplía aún más la región efectiva que cada activación resume.

---

## ¿Por qué la industria estandarizó la Opción B?

La industria favoreció arquitecturas tipo VGG y posteriores porque apilar filtros pequeños ofrece varias ventajas:

### 1. Más no linealidad

Tres capas convolucionales incorporan tres transformaciones y, usualmente, tres funciones de activación.  
Esto permite aprender patrones más complejos que una sola convolución grande.

### 2. Mejor extracción jerárquica de características

Las primeras capas pueden aprender bordes o texturas simples, mientras que las siguientes pueden combinar esas señales en patrones más abstractos.

### 3. Mayor control de la resolución espacial

El uso de filtros pequeños con padding permite conservar la dimensión espacial durante más etapas, lo cual da mayor control sobre cuándo reducir tamaño.

### 4. Mejor diseño para redes profundas

Aunque en este bloque particular la Opción B tiene más pesos, en arquitecturas profundas el uso repetido de filtros pequeños suele ser más flexible y más efectivo que depender constantemente de filtros grandes.

---

## Conclusión

La **Opción A** produce una salida de:

$$
\boxed{125 \times 125 \times 64}
$$

y requiere:

$$
\boxed{12544 \text{ pesos}}
$$

La **Opción B** produce una salida de:

$$
\boxed{128 \times 128 \times 64}
$$

y requiere:

$$
\boxed{76032 \text{ pesos}}
$$

Aunque la Opción B utiliza más parámetros en este caso puntual, su principal ventaja es que logra un **campo receptivo efectivo equivalente a $7 \times 7$** mediante capas pequeñas apiladas, incorporando más no linealidad, mejor jerarquía de características y un diseño más robusto.  
Por esa razón, la industria terminó estandarizando arquitecturas basadas en filtros pequeños apilados.

## Task 2 - Criterio

### 1. Justificación técnica

Los enfoques tradicionales como SIFT y HOG con SVM pueden funcionar en escenarios controlados, pero en un entorno agrícola real suelen volverse frágiles. En cultivos de mango, una misma enfermedad puede verse distinta según la hora del día, la intensidad de luz solar, la sombra de otras hojas, la humedad, el polvo, la inclinación de la cámara o la presencia de partes ocultas por ramas y frutos. SIFT y HOG dependen de descriptores manuales que capturan bordes, gradientes o puntos de interés locales, pero no comprenden de forma robusta el significado visual de la lesión. Por eso, cuando cambian las condiciones del entorno, esos descriptores pueden dejar de representar correctamente la enfermedad y el SVM termina generalizando peor fuera del conjunto de entrenamiento.

En cambio, una red profunda entrenada con **Transfer Learning** aprende **mapas de características** jerárquicos. En las primeras capas detecta patrones simples como bordes, texturas y contrastes; en capas intermedias combina esas señales para identificar formas, manchas, venaciones alteradas o patrones de color; y en capas profundas representa estructuras más semánticas relacionadas con el tipo de daño o enfermedad. Esto permite que el modelo no dependa únicamente de rasgos manuales rígidos, sino de representaciones aprendidas que toleran mejor variaciones de iluminación, perspectiva, escala y oclusión parcial. En términos prácticos, esto vuelve al sistema más robusto y más confiable para operar en condiciones reales de campo.

---

### 2. Elección entre AlexNet y VGG-16

Si la elección debiera hacerse estrictamente entre **AlexNet** y **VGG-16** para extraer características de hojas de mango, se elegiría **VGG-16**. La razón principal es que VGG-16 posee una arquitectura más profunda y modular, construida a partir de bloques repetidos de convoluciones pequeñas de $3 \times 3$. Esa modularidad permite aprender representaciones visuales de manera progresiva y ordenada: primero patrones locales, luego combinaciones de texturas y finalmente estructuras más abstractas asociadas a daños foliares complejos. Para una tarea como detección de enfermedades, donde pequeñas variaciones de color, textura y forma son importantes, esa profundidad adicional suele producir descriptores más ricos y discriminativos.

Además, VGG-16 incorpora repetidamente la función de activación **ReLU**, lo cual incrementa la no-linealidad de la red y le permite modelar relaciones visuales complejas que un diseño más superficial no captura con la misma precisión. Aunque AlexNet fue una arquitectura histórica muy importante, su capacidad de abstracción es menor en comparación con VGG-16. Por ello, si el objetivo es usar **Transfer Learning** para obtener características robustas y reutilizables en un problema agrícola real, VGG-16 resulta una opción más sólida por su mayor profundidad, mejor jerarquía de representación y capacidad de capturar patrones semánticos más finos.

```python
# Celda 1 - Imports
import os
import copy
import random
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display, Markdown
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models
from torchvision.models import VGG16_Weights
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("aryashah2k/mango-leaf-disease-dataset")
```

```python
# Celda 2 - Configuración general
import os
import random
import numpy as np
import torch

SEED = 42
BATCH_SIZE = 32
EPOCHS = 5
LR = 1e-3
NUM_WORKERS = 2
IMAGE_SIZE = 224

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# Posibles rutas donde puede estar el dataset
possible_paths = [
    r".\mango-leaf-disease-dataset\versions\1",
    r"mango-leaf-disease-dataset\versions\1",
    r".\versions\1",
    r".\dataset",
]

DATA_DIR = None
for path in possible_paths:
    abs_path = os.path.abspath(path)
    if os.path.exists(abs_path):
        DATA_DIR = abs_path
        break

if DATA_DIR is None:
    print("\nNo se encontró el dataset en las rutas probadas.")
    print("\nContenido del directorio actual:")
    print(os.listdir(os.getcwd()))
    raise FileNotFoundError(
        "No se encontró la carpeta del dataset. "
        "Verifica en qué carpeta está exactamente 'mango-leaf-disease-dataset'."
    )
```

```python
# Celda 3 - Transformaciones
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std = [0.229, 0.224, 0.225]

train_transform = transforms.Compose([
    transforms.RandomResizedCrop(IMAGE_SIZE, scale=(0.8, 1.0)),
    transforms.RandomRotation(20),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std),
])

eval_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std),
])
```

```python

imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std = [0.229, 0.224, 0.225]

train_transform = transforms.Compose([
    transforms.RandomResizedCrop(IMAGE_SIZE, scale=(0.8, 1.0)),
    transforms.RandomRotation(20),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std),
])

eval_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std),
])
```

```python
base_dataset = datasets.ImageFolder(root=DATA_DIR)

class_names = base_dataset.classes
num_classes = len(class_names)
targets = [label for _, label in base_dataset.samples]

print("Clases encontradas:", class_names)
print("Número de clases:", num_classes)
print("Número total de imágenes:", len(base_dataset))
```

```python
indices = np.arange(len(base_dataset))

train_idx, temp_idx = train_test_split(
    indices,
    test_size=0.30,
    stratify=targets,
    random_state=SEED
)

temp_targets = [targets[i] for i in temp_idx]

val_idx, test_idx = train_test_split(
    temp_idx,
    test_size=0.50,
    stratify=temp_targets,
    random_state=SEED
)

print("Train:", len(train_idx))
print("Val:", len(val_idx))
print("Test:", len(test_idx))
```

```python
train_dataset_full = datasets.ImageFolder(root=DATA_DIR, transform=train_transform)
eval_dataset_full = datasets.ImageFolder(root=DATA_DIR, transform=eval_transform)

train_dataset = Subset(train_dataset_full, train_idx)
val_dataset = Subset(eval_dataset_full, val_idx)
test_dataset = Subset(eval_dataset_full, test_idx)

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    pin_memory=torch.cuda.is_available()
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=torch.cuda.is_available()
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=torch.cuda.is_available()
)

print("Dataloaders listos.")
```

```python
images, labels = next(iter(train_loader))
print("Shape imágenes:", images.shape)
print("Shape etiquetas:", labels.shape)
print("Ejemplo etiquetas:", labels[:10].tolist())
```

```python
weights = VGG16_Weights.IMAGENET1K_V1
model = models.vgg16(weights=weights)

# Congelar extractor de características convolucional
for param in model.features.parameters():
    param.requires_grad = False

# También se congelan capas previas por seguridad
for param in model.parameters():
    param.requires_grad = False

# Reemplazar clasificador final
in_features = model.classifier[0].in_features

model.classifier = nn.Sequential(
    nn.Linear(in_features, 512),
    nn.ReLU(inplace=True),
    nn.Dropout(0.5),
    nn.Linear(512, 256),
    nn.ReLU(inplace=True),
    nn.Dropout(0.3),
    nn.Linear(256, num_classes)
)

# Solo las capas nuevas quedan entrenables
for param in model.classifier.parameters():
    param.requires_grad = True

model = model.to(device)

print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.classifier.parameters(), lr=LR)
```

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    all_preds = []
    all_labels = []

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        preds = torch.argmax(outputs, dim=1)

        all_preds.extend(preds.detach().cpu().numpy())
        all_labels.extend(labels.detach().cpu().numpy())

    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = accuracy_score(all_labels, all_preds)
    return epoch_loss, epoch_acc


def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)
            preds = torch.argmax(outputs, dim=1)

            all_preds.extend(preds.detach().cpu().numpy())
            all_labels.extend(labels.detach().cpu().numpy())

    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = accuracy_score(all_labels, all_preds)
    return epoch_loss, epoch_acc, all_labels, all_preds
```

```python
history = {
    "train_loss": [],
    "train_acc": [],
    "val_loss": [],
    "val_acc": []
}

best_val_loss = float("inf")
best_model_wts = copy.deepcopy(model.state_dict())

for epoch in range(EPOCHS):
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc, _, _ = evaluate(model, val_loader, criterion, device)

    history["train_loss"].append(train_loss)
    history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss)
    history["val_acc"].append(val_acc)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_model_wts = copy.deepcopy(model.state_dict())

    print(
        f"Época {epoch+1}/{EPOCHS} | "
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
    )

model.load_state_dict(best_model_wts)
print("Entrenamiento finalizado. Mejor modelo cargado.")
```

```python

completed_epochs = len(history["train_loss"])
epochs_range = range(1, completed_epochs + 1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, history["train_loss"], marker="o", label="Train Loss")
plt.plot(epochs_range, history["val_loss"], marker="o", label="Val Loss")
plt.xlabel("Época")
plt.ylabel("Loss")
plt.title("Curva de pérdida")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, history["train_acc"], marker="o", label="Train Acc")
plt.plot(epochs_range, history["val_acc"], marker="o", label="Val Acc")
plt.xlabel("Época")
plt.ylabel("Accuracy")
plt.title("Curva de accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
test_loss, test_acc, y_true, y_pred = evaluate(model, test_loader, criterion, device)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# Celda 14 - Matriz de confusión
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation="nearest", cmap="Blues")
plt.title("Matriz de confusión")
plt.colorbar()

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45, ha="right")
plt.yticks(tick_marks, class_names)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(
            j, i, str(cm[i, j]),
            horizontalalignment="center",
            color="white" if cm[i, j] > cm.max() / 2 else "black"
        )

plt.ylabel("Etiqueta real")
plt.xlabel("Etiqueta predicha")
plt.tight_layout()
plt.show()
```
